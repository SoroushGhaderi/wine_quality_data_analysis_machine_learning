```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red_wine = pd.read_csv("winequality-red.csv", sep= ";")
white_wine = pd.read_csv("winequality-white.csv", sep= ";")
```

```python
white_wine.head()
```

```python
red_wine.drop_duplicates(inplace= True)
white_wine.drop_duplicates(inplace= True)
```

```python
white_wine.info()
```

```python
red_wine["quality"].unique()
```

```python
white_wine["quality"].unique()
```

```python
red_wine["density"].mean()
```

```python
red_wine["frame"] = "Red"
```

```python
white_wine["frame"] = "White"
```

```python
white_wine
```

```python
wine = pd.concat([red_wine, white_wine], ignore_index= True, axis = 0)
```

```python
quality_group = wine.groupby(["quality", "frame"])
```

```python
quality_group.mean()
```

```python
wine.groupby("frame").mean()
```

```python
wine["pH"].describe()
```

```python
bin_edges = [2.72, 3.11, 3.21, 3.33, 4.01]
bin_names = ['high', 'higher_intermediate' , 'intermediate', "low"]
```

```python
wine["acidity_levels"] = pd.cut(wine["pH"], bin_edges, labels=bin_names)
```

```python
wine.head()
```

```python
wine.groupby("acidity_levels").mean()
```

```python
wine.groupby("frame")["quality"].mean().plot(kind= "bar", title = "Average Quality Of Wine", color = ["red", "blue"], alpha = .8)
plt.xlabel("Type of Wine", fontsize = 18)
plt.ylabel("Average Quality", fontsize = 18)
plt.show()
```

```python
wine.groupby("acidity_levels").plot()
```

```python
wine.head()
```

```python
wine["alcohol"].describe()
```

```python
wine_low_alcohol = wine.query('alcohol <  10.5')
wine_high_alcohol = wine.query('alcohol >=  10.5')
```

```python
wine_high_alcohol
```

```python
wine_low_alcohol
```

```python
alcohol_bins = [8, 10.4, 14.9]
alcohol_names = ["low", "high"]
wine["alcoholic_level"] = pd.cut(wine["alcohol"], alcohol_bins, labels= alcohol_names)
```

```python
wine.groupby("alcoholic_level").mean()
```

```python
wine.groupby(["alcoholic_level", "acidity_levels"]).mean()
```

```python
mean_quality_low = wine_low_alcohol["quality"].mean()
mean_quality_high = wine_high_alcohol["quality"].mean()
locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ["Low", "High"]
plt.bar(locations, heights, tick_label=labels)
plt.title("Average Quality Ratings by Alcohol Content")
plt.xlabel("Alcohol Content")
plt.ylabel("Average Quality Rating")
```

```python
fig, ax = plt.subplots(figsize = (12, 12))
df_corr = wine.corr()
ax = sns.heatmap(df_corr, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
